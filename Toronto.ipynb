```python
from bs4 import BeautifulSoup
```

```python
url = rq.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
soup = BeautifulSoup(url, 'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
```

```python
My_table
```

```python
table1 = soup.find_all('table',class_="wikitable")
```

```python
linhas = [item.get_text() for item in table1[0].find_all('td')]
linhas
```

```python
headers=['PostalCode', 'Borough', 'Neighborhood']
data = list(zip(*[iter(linhas)]*3))
Neighborhoods = pd.DataFrame(data[0:], columns=headers)
Neighborhoods['Neighborhood'] = Neighborhoods['Neighborhood'].str.rstrip('\n')
```

```python
Neighborhoods
```

```python
Clean = Neighborhoods.drop(Neighborhoods[(Neighborhoods.Borough=='Not assigned') & (Neighborhoods.Neighborhood=='Not assigned')].index)
```

```python
Clean= Clean.replace({'Neighborhood': 'Not assigned'}, "Queen's Park")
```

```python
Clean
```

```python
urlgeo='http://cocl.us/Geospatial_data'
LatLon = pd.read_csv(urlgeo)
LatLon.rename(columns={'Postal Code':'PostalCode'},inplace=True)
```

```python
LatLon
```

```python
Clean=Clean.merge(LatLon,on='PostalCode',how='right')
```

```python
Clean
```

```python
FinalVersion = Clean.groupby(['PostalCode','Borough','Latitude','Longitude'],sort=False,as_index=False).agg(','.join)
```

```python
FinalVersion=FinalVersion[['PostalCode','Borough','Neighborhood','Latitude','Longitude']]
FinalVersion
```

```python
FinalVersion.shape
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes
import folium

!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim
```

```python
address = 'Toronto, Ontario'

geolocator=Nominatim(user_agent='tor_explorer')
location=geolocator.geocode(address)
latitude=location.latitude
longitude=location.longitude
```

```python
toronto=folium.Map(location=[latitude,longitude],zoom_start=10)

for lat,lng,borough,neighborhood in zip(FinalVersion['Latitude'],FinalVersion['Longitude'],FinalVersion['Borough'],FinalVersion['Neighborhood']):
    label='{},{}'.format(neighborhood,borough)
    label=folium.Popup(label,parse_html=True)
    folium.CircleMarker([lat,lng],radius=5,popup=label,color='blue',fill=True,fill_color='#3186cc',fill_opacity=2,
                       parse_html=False).add_to(toronto)
    
toronto
```
